## Linear Regression is a statistical machine learning model used to predict a continuous target variable based on one or more input features.

```python
class LinReg():
        def _init_(self):
            self.m = None
            self.b = None

        def fit(self, X_train, Y_train):

            for i in range (X_train.shape[0]):
                num = 0 
                denum = 0
                num += ((X_train[i] - X_train.mean())*(Y_train[i] - Y_train.mean()))
                denum += (X_train[i] - X_train.mean())* (X_train[i] - X_train.mean())

                self.m = num/denum
                self.b = (Y_train.mean() - self.m * X_train.mean())

                print(self.m)
                print(self.b)
        def predict(self, X_test):
            print(X_test)
            return self.m * X_test + self.b
```

### In the train_test_split function from scikit-learn, the parameters test_size=0.2 and random_state=2 have the following meanings:

### test_size=0.2: This specifies the proportion of the dataset to include in the test split. Here, 0.2 means 20% of the data will be used for the test set (X_test and y_test), while the remaining 80% will be used for the training set (X_train and y_train).
### random_state=2: This sets a seed for the random number generator used to shuffle the data before splitting. By setting random_state=2, you ensure that the data is shuffled and split in the same way every time you run the code, making the results reproducible. The value 2 is arbitrary; any integer can be used as the seed.
